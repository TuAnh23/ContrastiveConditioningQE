```python
from utils import load_text_file
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
output_dir = "output/dev_knownMT/remove_token"

ok_prob_diff = load_text_file(f"{output_dir}/ok_prob_diff.txt")
bad_prob_diff = load_text_file(f"{output_dir}/bad_prob_diff.txt")

ok_prob_diff = [float(x) for x in ok_prob_diff]
bad_prob_diff = [float(x) for x in bad_prob_diff]
```

```python
# Combine the data for plotting
combined_data = [ok_prob_diff, bad_prob_diff]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(3,6))

# Create a box plot
box_plot = ax.boxplot(combined_data)

# Add labels and title
ax.set_xticklabels(['OK', 'BAD'])
ax.set_ylabel('Probability difference.')
ax.set_title('Probability difference \nusing contrastive conditioning.')

plt.savefig(f"{output_dir}/box_plot.png")

```

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Create kernel density plots using seaborn
sns.kdeplot(ok_prob_diff, ax=ax, label='OK')
sns.kdeplot(bad_prob_diff, ax=ax, label='BAD')

# Add labels and title
ax.set_xlabel('Probability difference')
ax.set_ylabel('')
ax.set_title('Probability difference using contrastive conditioning.')

# Add a legend
ax.legend()

plt.savefig(f"{output_dir}/density_graph.pdf")
```
